```python
### Import Required Libraries
import numpy as np
import pandas as pd
```

```python
### Import df
df = pd.read_csv('/content/drive/MyDrive/Open Data/equids-canton.csv', sep = ',')
```

```python
### Drop Depreciated Column
df.drop(['Month'], axis = 1, inplace = True)
```

```python
### Rename Column
df = df.rename(columns = {
    '??': 'NaN',
})
```

```python
### Compute Average per Year/Canton
df = df.groupby(['Year']).mean().round(0).astype(int)
```

```python
### Compute National Census
df['CH'] = df.sum(axis=1).astype(int)
```
